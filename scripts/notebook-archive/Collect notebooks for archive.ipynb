```python
import sys
import io
import os
import argparse
import nbformat

def remove_outputs(nb):
    for cell in nb.cells:
        if cell.cell_type == 'code':
            cell.outputs = []

def clear_notebook(old_ipynb, new_ipynb):
    with io.open(old_ipynb, 'r') as f:
        nb = nbformat.read(f, nbformat.NO_CONVERT)

    remove_outputs(nb)

    with io.open(new_ipynb, 'w', encoding='utf8') as f: 
        nbformat.write(nb, f, nbformat.NO_CONVERT)
```

```python
notebook_list = [
    'Collect notebooks for archive',
    'Collate OR estimates',
    'OR estimates',
    'Carrier tables for specific scores from Tejura',
    'Build exposure tables for Predictors for Chen et al',
    'Prepare environment',
    'OR estimates for predictors',
    'Prepare control pools for loading',
    'Case pools with time to diagnosis',
    'Build spliceAI person lists',
    'Per-gene VAT import',
    'README',
    'Build exposure tables for ClinVar 2025-01',
    'Build exposure tables for rare variants',
    'Build exposure tables for ClinVar (VAT)',
    'Make Long QT Case and Control Pools',
    'Load common tables',
    'Collect auxiliary variant statistics',
    'Prevalence curves',
    'Plot survival curves'
]
```

```python
write_to = 'notebook-archive'
from tqdm.notebook import tqdm
for notebook in tqdm(notebook_list):
    clear_notebook(f'{notebook}.ipynb', f'{write_to}/{notebook}.ipynb')
```
