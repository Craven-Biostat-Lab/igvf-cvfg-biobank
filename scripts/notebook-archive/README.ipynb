### CVFG dataframe

Switching to the CVFG dataframe as the source for a lot of information.

Note that as of this writing, the records are restricted, so a Zenodo access token and permission to view are required.

To get files into AoU from Zenodo, we use `curl` and the REST API.
Some relevant commands:

Listing files in a record:
```
curl -i https://zenodo.org/api/deposit/depositions/[RECORD]/files?access_token=[ACCESS_TOKEN]
```

Downloading a file:
```
curl -o [DESTINATION] [URL from the 'download' field of the listing]?access_token=[ACCESS_TOKEN]
```

Our workflow was to download the dataframe zip, unzip, gzip the csv file, and place it in the workspace bucket in the folder `cvfg_[RECORD]`, this way we track the version by the Zenodo record number.

### Organization as of 2025-02-07

The following notebooks need to run in the order below:

1. **`Per-gene VAT import`**: This takes the [AoU Variant Annotation Table (VAT)](https://support.researchallofus.org/hc/en-us/articles/4615256690836-Variant-Annotation-Table) and splits it up into per-gene files for our genes of interest and filters them to missense only. This should only be re-run if we are adding genes or if the VAT is updated. Review exclusion filter in the notebook before running again. Can be run in the background using `Run Per-gene VAT import in the Background`.
1. **`Missense counts for genes of interest`**: Loads case cohort data and saves it to gzipped `tsv` files for downstream analysis, and counts missense variants in the case cohorts. Should only be re-run if cohort definitions change.
1. **`Prepare control pools for loading`**: Loads control cohort data and saves it to gzipped `tsv` files for downstream analysis.
1. **`OR for MAVEs`**: Cmoputes ORs using Logistic Regression and saves results in gzipped json files. Can be run in the background using `Run MAVE OR in Background`. 
1. **`OR for MAVEs plot`**: Generates plots from the output of `OR for MAVEs`.
